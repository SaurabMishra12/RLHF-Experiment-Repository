# **Prompt Dataset**

```python
prompt_dataset_path = '/content/sample_prompt.jsonl'
```

```python
prompt_data = []
```

```python
len(prompt_data)
```

```python
def print_d(d):
    for key, val in d.items():
        print(f"key:{key}\nval:{val}\n")
```

```python
print_d(prompt_data[0])
```
